1. Read the data set and replace dashes with 0 to make sure you can perform arithmetic operations on the data.

4. Write a function with name â€œGoal_diff_countâ€ which should return all the teams with their Goal Differences. Using the same function, find the team which has maximum and minimum goal difference.

```python
#Write a function with name â€œGoal_diff_countâ€ which should return all the teams with their Goal Differences.
def Goal_diff_count(a,b):
    diff1 = a - b
    return diff1

ftball_df['Goal Diff'] = Goal_diff_count(ftball_df['GoalsFor'].astype(float),ftball_df['GoalsAgainst'].astype(float))

print('The team with maximum goal difference is: ')
print(ftball_df[['Team', 'Goal Diff']][ftball_df['Goal Diff']==ftball_df['Goal Diff'].max()])

print()
print('The team with minimum goal difference is: ')
print(ftball_df[['Team', 'Goal Diff']][ftball_df['Goal Diff']==ftball_df['Goal Diff'].min()])
```

5. Create a new column with name â€œWinning Percentâ€ and append it to the data set. 
   Percentage of Winning = (GamesWon / GamesPlayed)*100
   If there are any numerical error, replace it with 0%
   Print the top 5 teams which has the highest Winning percentage

```python
#Create a new column with name â€œWinning Percentâ€ and append it to the data set. 
ftball_df['GamesWon'] = ftball_df['GamesWon'].astype(int)
ftball_df['GamesPlayed'] = ftball_df['GamesPlayed'].astype(int)

ftball_df['Winning Percent'] = (ftball_df['GamesWon'] / ftball_df['GamesPlayed'])*100
ftball_df['Winning Percent']= ftball_df['Winning Percent'].fillna(0).round(2).astype(str) + '%'
ftball_df

#Print the top 5 teams which has the highest Winning percentage
ftball_df['Winning Percent'].max()
ftball_df.groupby('Winning Percent').head(5).reset_index(drop=True)
ftball_df.head()
```
